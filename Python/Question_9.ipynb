```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Reading the data from the following link,
## https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD

```python
url = 'https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD'
df = pd.read_csv(url)
df.head()
```

# Get all the cars and their types that do not qualify for clean alternative fuel vehicle


```python
non_clean_fuel_cars = df[df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] == 'Not eligible due to low battery range']
non_clean_fuel_cars_info = non_clean_fuel_cars[['Make', 'Model']]
print("Cars and their types that do not qualify for clean alternative fuel vehicle:")
print(non_clean_fuel_cars_info)
```

# Get all TESLA cars with the model year, and model type made in Bothell City.


```python
tesla_made_in_bothell = df[(df.Make == 'TESLA') & (df.City == 'BOTHELL')]
print("Tesla cars made in Bothell city:")
tesla_made_in_bothell[['Make','Model','City','Model Year']]
```

# Get all the cars that have an electric range of more than 100, and were made after 2015


```python
made_after_2015 = df[(df['Model Year']>2015) & (df['Electric Range']>100)]
print("Cars that have an electric range of more than 100, and were made after 2015")
made_after_2015[['Make','Electric Range','Model Year']]
```
