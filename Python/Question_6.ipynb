```python
import pandas as pd
import requests
import numpy as np
from datetime import datetime
```

```python
url = "https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json"
res = requests.get(url)
data = res.json()
df = pd.DataFrame(data['pokemon'])
df.head()
```

# Pokemons whose spawn rate is less than 5%

```python
spawn_per_lt_5 = df[df.spawn_chance<5]
print(len(spawn_per_lt_5))
spawn_per_lt_5[['name','spawn_chance']]
```

# Pokemons that have less than 4 weaknesses

```python
df['weaknesses_length']=df["weaknesses"].apply(len)
weaknesses_lt_4 = df[df.weaknesses_length<4]
len(weaknesses_lt_4)
weaknesses_lt_4[['name','weaknesses_length']]
```

# Pokemons that have no multipliers at all

```python
no_multipliers = df[df.multipliers.isnull()]
print(len(no_multipliers))
no_multipliers['name']
```

# Pokemons that do not have more than 2 evolutions

```python
mt_2_evolutions = df[df.next_evolution.str.len()<=2]
print(len(mt_2_evolutions))
mt_2_evolutions[['name','next_evolution']]
```

# Pokemons whose spawn time is less than 300 seconds.


```python
spawn_time_less_than_300 = df[df["spawn_time"].replace("N/A", "24:00").apply(lambda x: int(x.split(":")[0]) * 60 + int(x.split(":")[1])) < 300]
print(len(spawn_time_less_than_300))
spawn_time_less_than_300[['name','spawn_time']]
```

# Pokemon who have more than two types of capabilities

```python
mt_2_capabilities = df[df.type.apply(len)>2]
print(len(mt_2_capabilities))
mt_2_capabilities[['name','type']]
```
